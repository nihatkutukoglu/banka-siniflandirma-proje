# Banka Pazarlama Veri Seti Sütun Açıklamaları

Aşağıda veri setimizde bulunan özelliklerin (sütunların) isimleri, veri tipleri ve anlamları listelenmiştir:

### 👤 Müşteri Bilgileri (Kişisel Veriler)
* **`age`** (Yaş): Müşterinin yaşı *(Sayısal)*.
* **`job`** (Meslek): Müşterinin çalışma türü / mesleği *(Kategorik: "admin.", "blue-collar", "entrepreneur", "housemaid", "management", "retired", "self-employed", "services", "student", "technician", "unemployed", "unknown")*.
* **`marital`** (Medeni Durum): Müşterinin medeni hali *(Kategorik: "divorced" (boşanmış/dul), "married" (evli), "single" (bekar))*.
* **`education`** (Eğitim): Müşterinin eğitim seviyesi *(Kategorik: "primary" (ilköğretim), "secondary" (ortaöğretim), "tertiary" (yükseköğretim), "unknown" (bilinmiyor))*.

### 💳 Müşterinin Banka ve Kredi Geçmişi
* **`default`** (Kredi Temerrüdü): Müşterinin geçmişte ödenmemiş borcu (kredi temerrüdü) var mı? *(Kategorik: "yes" (evet), "no" (hayır))*.
* **`balance`** (Bakiye): Müşterinin banka hesabındaki ortalama yıllık bakiye (Euro/Dolar vb. cinsinden) *(Sayısal)*.
* **`housing`** (Ev Kredisi): Müşterinin ev kredisi var mı? *(Kategorik: "yes" (evet), "no" (hayır))*.
* **`loan`** (Bireysel Kredi): Müşterinin bireysel kredisi var mı? *(Kategorik: "yes" (evet), "no" (hayır))*.

### 📞 Mevcut Kampanya ile İlgili İletişim Bilgileri
* **`contact`** (İletişim Türü): Müşteriyle hangi yöntemle iletişim kurulduğu *(Kategorik: "cellular" (cep telefonu), "telephone" (sabit hat), "unknown" (bilinmiyor))*.
* **`day`** (Gün): Müşteriyle ayın kaçıncı gününde iletişim kurulduğu (1-31 arası) *(Sayısal)*.
* **`month`** (Ay): Müşteriyle yılın hangi ayında iletişim kurulduğu *(Kategorik: "jan", "feb", "mar" ... "nov", "dec")*.
* **`duration`** (Süre): Son iletişimin (telefon görüşmesinin) saniye cinsinden süresi *(Sayısal)*. **Önemli Not:** Bu özellik, modelin tahmin gücünü çok yüksek oranda etkiler. Eğer `duration = 0` ise müşteri hesabını açmamıştır (`deposit = no`). Gerçek dünya uygulamalarında, görüşme bitmeden bu süre bilinemeyeceği için tahmin modellemesinde dikkatli kullanılmalı veya çıkarılmalıdır.
* **`campaign`** (Kampanya İletişim Sayısı): Bu kampanya özelinde müşteriyle kaç kere iletişime geçildiği *(Sayısal, son iletişim dahil)*.

### 🔄 Önceki Kampanya Geçmişi
* **`pdays`** (Geçen Gün Sayısı): Bir önceki kampanyada müşteriyle iletişim kurulduktan sonra geçen gün sayısı *(Sayısal: `-1` değeri müşteriyle daha önce hiç iletişim kurulmadığını gösterir)*.
* **`previous`** (Önceki İletişim Sayısı): Bu kampanyadan önce müşteriyle gerçekleştirilen iletişim sayısı *(Sayısal)*.
* **`poutcome`** (Önceki Kampanya Sonucu): Bir önceki pazarlama kampanyasının sonucu *(Kategorik: "failure" (başarısız), "success" (başarılı), "other" (diğer), "unknown" (bilinmiyor))*.

### 🎯 Hedef Değişken (Çıktı)
* **`deposit`** (Vadeli Mevduat): Müşteri vadeli mevduat (term deposit) hesabı açtırdı mı? *(Kategorik: "yes" (evet açtırdı), "no" (hayır açtırmadı))*.

# 🧹 BÖLÜM 2: Veri Temizliği (Data Cleaning)

Keşifçi Veri Analizi'ne (EDA) geçmeden ve modelleri eğitmeden önce, veri setimizi algoritmaların anlayabileceği kadar pürüzsüz ve "sağlıklı" hale getirmemiz gerekir. Çünkü makine öğrenmesinin altın kuralı şudur: **"Garbage In, Garbage Out" (Çöp girerse, çöp çıkar).**

Veri setini incelediğimizde kullanıcı kaynaklı veri giriş hataları, kopya satırlar, eksik (NaN) değerler, yazım yanlışları ve mantıksız aykırı rakamlar (Outliers) tespit ettik.

Bu bölümde sırasıyla şu operasyonları gerçekleştireceğiz:
1. **Kopya Verilerin Silinmesi:** Veri tabanına yanlışlıkla çift kaydedilmiş birebir aynı satırların (`duplicates`) temizlenmesi.
2. **Yazım Hatalarının Düzeltilmesi:** `balance` ve `marital` sütunlarına manuel yazılmış hatalı/tutarsız metinlerin ("Soylemedi", "EVLİ", "singel") düzeltilmesi.
3. **Eksik ve Gizli Verilerin Doldurulması:** Gerçek boşlukların (NaN) istatistiksel metotlarla (Medyan/Mod) doldurulması ve bazı sütunlardaki anlamsız "unknown" ibarelerinin makine öğrenmesine uygun şekilde yeniden etiketlenmesi.
4. **İmkansız Aykırı Değerlerin Silinmesi:** İnsan biyolojisine aykırı olan yaş verilerinin (Örn: Yaş = 3000) veri setinden atılması.


```python
# 9. Veri Temizliği: Tekrar Eden Satırları Silme
print("##################### Veri Temizliği: Duplicates #####################")

# Silme işleminden önceki satır sayısını alıyoruz
onceki_satir_sayisi = df.shape[0]

# drop_duplicates() ile birebir aynı olan kopyaları siliyoruz.
# inplace=True parametresi, değişikliği doğrudan mevcut 'df' içine kalıcı olarak kaydeder.
df.drop_duplicates(inplace=True)

# Silme işleminden sonraki satır sayısı
sonraki_satir_sayisi = df.shape[0]
silinen_satir = onceki_satir_sayisi - sonraki_satir_sayisi

print(f"İşlem tamam! Başarıyla silinen kopya satır sayısı: {silinen_satir}")
print(f"Güncel veri seti boyutumuz (Satır, Sütun): {df.shape}")
print("-----------------------------------------------------------\n")
```

```python
# 1. Adım: 'balance' sütununu tamamen sayısal (numeric) bir sütuna dönüştürüyoruz.
# Bu işlem sayesinde 'Soylemedi' gibi metin olan tüm hatalı veriler otomatik olarak 'NaN' (boş değer) e dönüşecek.
df['balance'] = pd.to_numeric(df['balance'], errors='coerce')

# 2. Adım: 'balance' sütununun medyanını hesaplıyoruz (NaN olanlar hesaplamaya katılmaz)
medyan_bakiye = df['balance'].median()

# 3. Adım: NaN (boş) olan yerleri (yani eski 'Soylemedi' değerini) hesapladığımız medyan ile dolduruyoruz.
df['balance'] = df['balance'].fillna(medyan_bakiye)

print(f"İşlem Başarılı! Hatalı 'Soylemedi' değeri silindi ve yerine medyan olan {medyan_bakiye} yazıldı.")

# İstersen kontrol etmek için null değer sayısına tekrar bakabilirsin:
# print("Kalan boş değer sayısı:", df['balance'].isnull().sum())
```

```python
# 1. Adım: 'marital' sütunundaki en çok tekrar eden değeri (mod) buluyoruz
en_cok_tekrar_eden_durum = df['marital'].mode()[0]

# 2. Adım: Hatalı olan 'EVLİ' ve 'singel' değerlerini doğrudan bulduğumuz mod değeri ile değiştiriyoruz
df['marital'] = df['marital'].replace(['EVLİ', 'singel'], en_cok_tekrar_eden_durum)

print(f"İşlem Başarılı! Hatalı kelimeler silindi ve yerine en çok tekrar eden '{en_cok_tekrar_eden_durum}' yazıldı.")

# Son durumu kontrol etmek için eşsiz değerleri tekrar ekrana yazdıralım:
print("\nGüncel 'marital' sütunu değerleri:")
print(df['marital'].unique())
```

```python
# 1. GERÇEK BOŞLUKLARI (NaN) DOLDURMA
df['age'] = df['age'].fillna(df['age'].median())
df['education'] = df['education'].fillna(df['education'].mode()[0])
df['housing'] = df['housing'].fillna(df['housing'].mode()[0])
print("✔️ Gerçek boş değerler (NaN) başarıyla dolduruldu.")
```

```python
# 2. 'UNKNOWN' DEĞERLERİ YÖNETME
# Az sayıda olan job ve education sütunlarındaki unknown'ları mod ile değiştiriyoruz
df['job'] = df['job'].replace('unknown', df['job'].mode()[0])
df['education'] = df['education'].replace('unknown', df['education'].mode()[0])
```

```python
# Verinin %75'i olan poutcome sütununu silmek yerine anlamlı bir isme çeviriyoruz
df['poutcome'] = df['poutcome'].replace('unknown', 'Ilk_Kez_Araniyor')
print("✔️ 'unknown' gizli boşlukları başarıyla tedavi edildi.")
```

### 6. Aykırı Değer (Outlier) Analizi ve Temizliği
Makine öğrenmesi modelleri aşırı uç değerlere (outliers) karşı hassastır. Öncelikle sayısal sütunlarımızdaki (`age`, `balance`, `duration`) aykırı değerleri bir **Kutu Grafiği (Boxplot)** ile görselleştirelim. Ardından mantıksız olan hatalı girişleri verimizden silelim.

```python
# 1. Kutu Grafiği (Boxplot) matematiğine göre Üst Sınırı (Outlier sınırını) bulalım:
Q1 = df['duration'].quantile(0.25)
Q3 = df['duration'].quantile(0.75)
IQR = Q3 - Q1
ust_sinir = Q3 + 1.5 * IQR

print(f"📊 İstatistiksel olarak Süre (Duration) için Aykırı Değer (Outlier) Üst Sınırı: {ust_sinir:.0f} saniye")

# 2. Bu sınırın üzerindeki (outlier olan) çok uzun çağrıları filtreleyelim
outlier_musteriler = df[df['duration'] > ust_sinir]

# 3. Bu outlier müşteriler hesabı açtırmış mı? (Oranlarına bakalım)
# Not: deposit sütunu ön işleme bağlı olarak 1/0 veya yes/no olabilir.
outlier_basari_orani = outlier_musteriler['deposit'].value_counts(normalize=True) * 100

print(f"\n📞 Toplam {len(outlier_musteriler)} adet aykırı değer (çok uzun) telefon görüşmesi var.")
print("Bu ÇOK UZUN görüşmelerin başarı (Hesap Açtırma) oranı:")
print(outlier_basari_orani.round(2))

# 4. Kıyaslamak için NORMAL süreli görüşmelerin başarısına da bakalım:
normal_musteriler = df[df['duration'] <= ust_sinir]
normal_basari_orani = normal_musteriler['deposit'].value_counts(normalize=True) * 100

print("\n⏱️ Normal süreli görüşmelerin başarı (Hesap Açtırma) oranı:")
print(normal_basari_orani.round(2))
```

### 🔍 Neden Eksik Verileri Her Zaman "Mod" İle Doldurmamalıyız? (Verinin Sırrı)

Keşifsel Veri Analizi (EDA) aşamasında `contact` (iletişim türü) sütununda çok yüksek oranda **"unknown"** (bilinmeyen/belirtilmemiş) veri olduğunu tespit ettik. Normal şartlarda makine öğrenmesinde eksik kategorik veriler, en çok tekrar eden değer (Mod) ile doldurulabilir. 

> 💡 **Ancak biz bu veri setinde bunu bilerek yapmadık ve "unknown" değerlerini ayrı bir kategori olarak bıraktık.** Peki neden?

Çünkü bu bankacılık verisinde iletişim bilgisinin "bilinmiyor" olması rastgele bir kayıt hatası değildir; aslında altında çok önemli bir iş bilgisi yatar. Eğer banka bir müşterinin cep telefonu *(cellular)* veya sabit hat *(telephone)* bilgisini bilmiyorsa, o müşteri büyük ihtimalle yıllardır bankaya uğramamış, bilgilerini güncellememiş **"pasif"** bir müşteridir. 

🎯 **Analitik Hipotezimiz:**
**İletişim bilgisi bilinmeyen (pasif) müşterilerin kampanyayı kabul edip vadeli hesap açtırma ihtimali, cep telefonu bilinen (aktif) müşterilere göre çok daha düşük olmalıdır.**

Aşağıdaki grafikte tam olarak bu hipotezin veriyle nasıl doğrulandığını göreceğiz. 👇


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. GRAFİK İLE GÖRSEL KANIT
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# İletişim türüne göre hesap açtırma durumunu görselleştiriyoruz
sns.countplot(data=df, x='contact', hue='deposit', palette=['#ff6b6b','#4ecdc4'])

plt.title("İletişim Türüne Göre Vadeli Hesap Açtırma Dağılımı", fontsize=14, fontweight='bold')
plt.xlabel("İletişim Türü (Contact)", fontsize=12)
plt.ylabel("Müşteri Sayısı", fontsize=12)
plt.legend(title='Hesap Açtırdı mı? (Deposit)')

plt.show()

# 2. RAKAMLARLA İSTATİSTİKSEL KANIT (Çapraz Tablo)
print("\n################# İSTATİSTİKSEL KANIT (BAŞARI YÜZDELERİ) #################")
print("Her bir iletişim türündeki müşterilerin yüzde kaçı teklifi kabul etmiş?\n")

# Çapraz tablo oluşturup satır bazında (index) yüzdelerini alıyoruz
oranlar = pd.crosstab(df['contact'], df['deposit'], normalize='index') * 100

# Formatlayıp ekrana yazdırıyoruz
for iletisim_turu in oranlar.index:
    kabul_orani = oranlar.loc[iletisim_turu, 'yes']
    red_orani = oranlar.loc[iletisim_turu, 'no']
    print(f"👉 {iletisim_turu.upper()} olanların: %{kabul_orani:.2f}'si 'EVET', %{red_orani:.2f}'si 'HAYIR' demiş.")
```

### 💡 Analiz Sonucu ve Çıkarımlar

Yukarıdaki grafik ve yüzdelik oranlar, başlangıçtaki hipotezimizi çok net bir şekilde doğrulamaktadır:

1. **"Unknown" (Pasif) Müşteri Davranışı:** İletişim bilgisi bankada kayıtlı olmayan (unknown/belirtilmemiş) müşterilerin teklifi reddetme (HAYIR) oranı devasa boyutlardadır. Bu grubun kampanya teklifini kabul etme ihtimali diğerlerine kıyasla çok düşüktür. Bu tablo, iletişim bilgisi eksik olan kişilerin bankayla bağlarının zayıf olduğu teorimizi istatistiksel olarak kanıtlamaktadır.
2. **"Cellular" (Aktif) Müşteri Davranışı:** Buna karşılık, bankaya cep telefonu bilgisini vermiş olan, profili güncel ve doğrudan ulaşılabilen müşterilerde kampanyaya katılım (EVET) oranı dramatik bir şekilde yüksektir.
3. **Model Başarısına Etkisi:** Eğer veri temizliği aşamasında "unknown" değerlerini en çok tekrar eden (mod) değer olan "cellular" ile doldurduk.

### 💡 Çıkarım: Görüşme Süresi (Duration) ve Başarı Arasındaki Güçlü Bağ

Yukarıdaki Kutu Grafiği (Boxplot) incelendiğinde, müşteri temsilcisinin telefonda kalma süresi ile kampanyanın başarısı arasında **doğrudan ve çok güçlü bir ilişki** olduğu görülmektedir:

### 💎 Özel İş Zekası (BI) Analizi: VIP Müşteri Segmentasyonu

Keşifsel Veri Analizi (EDA) sürecimizi standart grafiklerin bir adım ötesine taşıyarak doğrudan **iş değeri (business value)** üretecek bir analize çeviriyoruz. Bir banka için pazarlama bütçesini doğru yönetmenin en iyi yolu, en kârlı hedef kitleyi belirlemektir.

Bu aşamada özel bir kural seti (Rule-based Segmentation) oluşturuyoruz:
* 💰 **VIP Kuralı:** Banka bakiyesi (balance) en yüksek **%10'luk dilimde** olan ve geçmişte hiç kredi borcunu aksatmamış (`default = 'no'`) müşterileri **"VIP Müşteri"** olarak etiketleyeceğiz. Geri kalanları ise "Standart Müşteri" olarak kabul edeceğiz.

**Bu Analizde Neyi Göreceğiz?**
1. VIP müşterilerin kampanyaya olumlu yanıt verme (vadeli hesap açtırma) ihtimali, standart müşterilere göre ne kadar yüksek?
2. Bankamızın en zengin %10'luk kesimi en çok hangi meslek gruplarından oluşuyor?

Aşağıdaki kodlarda önce bu matematiksel kuralı veri setimize uygulayacak, ardından sonuçları yan yana iki grafik ile görselleştireceğiz. 👇


```python
# 1. MATEMATİKSEL VIP KURALI (Top %10'luk Dilim)
# Veri setindeki bakiyelerin en yüksek %10'luk kısmını kesen para limitini (Quantile) buluyoruz.
vip_para_limiti = df['balance'].quantile(0.90)
# VIP Şartı: Bakiyesi en zengin %10'luk dilimde olanlar VE temerrüdü (borcu) olmayanlar
vip_sartlari = (df['balance'] >= vip_para_limiti) & (df['default'] == 'no')
df['Musteri_Tipi'] = np.where(vip_sartlari, 'VIP Müşteri', 'Standart Müşteri')
```

```python
# 2. GÖRSELLEŞTİRME (Yanyana 2 Grafik)
plt.figure(figsize=(15, 6))
# Birinci Grafik: VIP vs Standart Müşterilerin Kampanyaya Tepkisi
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Musteri_Tipi', hue='deposit', palette=['#ff6b6b','#4ecdc4'])
plt.title("VIP ve Standart Müşterilerin Kampanya Başarısı", fontsize=13, fontweight='bold')
plt.xlabel("Müşteri Segmenti", fontsize=11)
plt.ylabel("Kişi Sayısı", fontsize=11)
plt.legend(title='Hesap Açtırdı mı?')
# İkinci Grafik: Sadece VIP Müşterilerin Meslekleri
plt.subplot(1, 2, 2)
vip_df = df[df['Musteri_Tipi'] == 'VIP Müşteri'] # Sadece VIP'leri filtreledik
sns.countplot(data=vip_df, y='job', order=vip_df['job'].value_counts().index, palette='viridis')
plt.title("VIP Müşteriler En Çok Hangi Meslekten?", fontsize=13, fontweight='bold')
plt.xlabel("VIP Kişi Sayısı", fontsize=11)
plt.ylabel("Meslekler (Job)", fontsize=11)
plt.tight_layout()
plt.show()
```

```python
 #. İSTATİSTİKLERİ EKRANA YAZDIRMA
print("################ VIP MÜŞTERİ ANALİZ RAPORU ################")
print(f"Toplam Müşteri Sayısı: {len(df)}")
print(f"VIP Müşteri Sayısı: {len(vip_df)} (Tüm müşterilerin yaklaşık %{(len(vip_df)/len(df))*100:.1f}'i)")
# VIP ve Standart başarı oranlarını hesaplama
vip_kabul_orani = (vip_df[vip_df['deposit'] == 'yes'].shape[0] / len(vip_df)) * 100
standart_df = df[df['Musteri_Tipi'] == 'Standart Müşteri']
std_kabul_orani = (standart_df[standart_df['deposit'] == 'yes'].shape[0] / len(standart_df)) * 100
print(f"\n👉 VIP Müşterilerin Teklifi Kabul Etme Oranı: %{vip_kabul_orani:.1f}")
print(f"👉 Standart Müşterilerin Teklifi Kabul Etme Oranı: %{std_kabul_orani:.1f}")
```

### 💰 Bakiye (Balance) ile Kampanya Başarısı Arasındaki İlişki

1. **Yüksek Bakiyenin Gücü:** Teklifi kabul edip hesap açtıranların (kırmızı `yes` kutusu) **medyan (ortanca)** bakiyesinin, reddedenlere (mavi `no` kutusu) göre bariz şekilde daha yüksek olduğunu görüyoruz (kutuların içindeki yatay çizgiler).

3. **Boş Hesapların Reddetme Eğilimi:** Teklifi reddedenlerin (`no`) kutusu çok daha dar ve alt sınırlara (sıfıra doğru) yığılmış durumda. Yani hesabında parası olmayan müşteriler tahmin edeceğimiz üzere büyük oranda "Hayır" demeye eğilimliler.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

# Kutu Grafiğini Çizdiriyoruz
sns.boxplot(data=df, x='deposit', y='balance', palette=['#ff9999','#66b3ff'])

# Grafik okunabilir olsun diye Y eksenine (Bakiye) Zom (Limit) atıyoruz
plt.ylim(-1000, 8000) 

plt.title("Banka Bakiyesi ile Vadeli Hesap Açtırma Arasındaki İlişki", fontsize=14, fontweight='bold')
plt.xlabel("Hesap Açtırdı mı? (Deposit)", fontsize=12)
plt.ylabel("Banka Bakiyesi", fontsize=12)
plt.show()

# İSTATİSTİKSEL KANIT
print("--- İSTATİSTİKSEL KANIT (MEDYAN/ORTANCA BAKİYELER) ---")
kabul_eden_medyan = df[df['deposit'] == 'yes']['balance'].median()
reddeden_medyan = df[df['deposit'] == 'no']['balance'].median()

print(f"👉 'HAYIR' diyenlerin Medyan Bakiyesi: {reddeden_medyan}")
print(f"👉 'EVET' diyenlerin Medyan Bakiyesi: {kabul_eden_medyan}")
```

### 🕵️‍♂️ Verimsiz Pazarlama Eforu Analizi: Parasız Müşterileri Neden Arıyoruz?

Önceki grafikte "parası olanların daha çok yatırım yaptığını" kanıtladık. Akla hemen şu harika soru geliyor: **"Peki banka, bakiyesi SIFIR (0) veya eksiye düşmüş müşterileri arayarak boşuna zaman ve para mı kaybediyor?"**

Çağrı merkezinde yapılan her bir aramanın bankaya bir personel ve zaman maliyeti vardır. Acaba hedef kitlemizi yanlış mı seçiyoruz? Bunu test etmek için müşterileri 3 farklı bakiye segmentine ayıracağız:
1. **Sıfır veya Eksi Bakiye** (Hesabı tamamen boş veya ekside olanlar)
2. **Düşük Bakiye** (Hesabında 0 ile 1000 birim arası parası olanlar)
3. **Yüksek Bakiye** (Hesabında 1000 birimden fazla parası olanlar)

Gelin bankanın kimi aradığını ve bu aramaların başarı oranlarını şeffaf bir şekilde görelim:


```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bakiye Segmentlerini Oluşturan Bir Fonksiyon Yazıyoruz
def bakiye_segmentle(bakiye):
    if bakiye <= 0:
        return 'Sıfır veya Eksi Bakiye'
    elif bakiye < 1000:
        return 'Düşük Bakiye (0-1000)'
    else:
        return 'Yüksek Bakiye (1000+)'

# Segmentleri veri setimize yeni bir sütun olarak ekliyoruz
df['Bakiye_Segmenti'] = df['balance'].apply(bakiye_segmentle)

# 2. GÖRSELLEŞTİRME KISMI
plt.figure(figsize=(15, 6))

# Birinci Grafik: Kimi Arıyoruz? (Dağılım)
plt.subplot(1, 2, 1)
ax = sns.countplot(data=df, x='Bakiye_Segmenti', palette='Reds_r', 
                   order=['Sıfır veya Eksi Bakiye', 'Düşük Bakiye (0-1000)', 'Yüksek Bakiye (1000+)'])
plt.title("Banka En Çok Kimi Aramış? (Efor Dağılımı)", fontsize=13, fontweight='bold')
plt.xlabel("Müşteri Segmenleri", fontsize=11)
plt.ylabel("Aranan Kişi Sayısı", fontsize=11)

# İkinci Grafik: Dönüşüm Oranları (Başarı)
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Bakiye_Segmenti', hue='deposit', palette=['#ff9999','#66b3ff'],
              order=['Sıfır veya Eksi Bakiye', 'Düşük Bakiye (0-1000)', 'Yüksek Bakiye (1000+)'])
plt.title("Hangi Segment Daha Çok 'EVET' Dedi?", fontsize=13, fontweight='bold')
plt.xlabel("Müşteri Segmentleri", fontsize=11)
plt.ylabel("Kişi Sayısı", fontsize=11)
plt.legend(title='Hesap Açtırdı mı?')

plt.tight_layout()
plt.show()

# 3. İSTATİSTİKSEL KANITLARI YAZDIRMA
sifir_bakiye_df = df[df['Bakiye_Segmenti'] == 'Sıfır veya Eksi Bakiye']
sifir_kabul_sayisi = sifir_bakiye_df[sifir_bakiye_df['deposit'] == 'yes'].shape[0]
sifir_kabul_orani = (sifir_kabul_sayisi / len(sifir_bakiye_df)) * 100

print("############### ZAMAN & MALİYET İSRAFI RAPORU ###############")
print(f"Toplam Aranan Müşteri: {len(df)}")
print(f"Bakiyesi 'SIFIR veya EKSİ' Olmasına Rağmen Aranan Kişi Sayısı: {len(sifir_bakiye_df)}")
print(f"Zaman Kaybı Oranı: Bankanın yaptığı aramaların yaklaşık %{(len(sifir_bakiye_df)/len(df))*100:.1f}'i boş hesaplara yapılmış!")
print(f"Bu kitlede 'EVET' diyerek hesap açtıranların oranı sadece: %{sifir_kabul_orani:.1f}")
```

Bu aşamada `month` (ay) sütununu kullanarak iki kritik metrik ölçeceğiz:
1. **Bankanın Eforu:** Hangi ayda ne kadar çok arama/iletişim yapılmış?
2. **Bankanın Verimi:** Hangi ayda yapılan aramalar oransal olarak daha yüksek başarı (Dönüşüm / Conversion) getirmiş?

💡 **İş Zekası (BI) Hipotezimiz:** Bankanın en çok arama yaptığı "en yoğun" ay ile oransal olarak en çok 'EVET' yanıtı aldığı "en verimli" ay aynı olmayabilir! Eğer böyle bir durum varsa, banka çağrı merkezi gücünü yanlış aylarda tüketiyor demektir. Gelin bunu test edelim. 👇


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Grafiğin düzgün okunması için ayları kronolojik sıraya koyuyoruz
ay_siralamasi = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

# 1. GÖRSELLEŞTİRME
plt.figure(figsize=(15, 6))

# Birinci Grafik: Hangi ay ne kadar arama yapılmış? (Efor)
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='month', order=ay_siralamasi, palette='viridis')
plt.title("Aylara Göre Toplam Arama Sayısı (Efor)", fontsize=13, fontweight='bold')
plt.ylabel("Aranan Kişi Sayısı", fontsize=11)
plt.xlabel("Aylar", fontsize=11)

# İkinci Grafik: Hangi ay ne kadar EVET/HAYIR alınmış? (Sonuç)
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='month', hue='deposit', order=ay_siralamasi, palette=['#ff9999','#66b3ff'])
plt.title("Aylara Göre Kampanya Başarısı (Kabul/Ret)", fontsize=13, fontweight='bold')
plt.ylabel("Kişi Sayısı", fontsize=11)
plt.xlabel("Aylar", fontsize=11)
plt.legend(title='Hesap Açtırdı mı?')

plt.tight_layout()
plt.show()


# 2. İSTATİSTİKSEL KANIT (Hangi Ayın "Oranı" Daha Yüksek?)
# Sadece sayıca çok olan değil, "arama başına evet alma" oranı yüksek olan ayı buluyoruz
aylik_toplam = df.groupby('month').size()
aylik_evet = df[df['deposit'] == 'yes'].groupby('month').size()

# Her ayın başarı oranını yüzde olarak hesaplıyoruz
aylik_basari = (aylik_evet / aylik_toplam * 100).reindex(ay_siralamasi)

print("############### AYLIK SEZONSALLIK VE VERİM RAPORU ###############")
print("Aranan müşterilerin yüzde kaçı teklifi kabul etti?")
for ay, oran in aylik_basari.items():
    if not pd.isna(oran):
        print(f"- {ay.upper()} Ayı: %{oran:.1f} Başarı")

en_iyi_ay = aylik_basari.idxmax()
en_yuksek_oran = aylik_basari.max()

en_cok_aranan_ay = aylik_toplam.idxmax()

print("-" * 65)
print(f"📞 EN ÇOK ARAMA YAPILAN (EN YORUCU) AY: {en_cok_aranan_ay.upper()}")
print(f"🏆 EN VERİMLİ (ORANSAL OLARAK EN BAŞARILI) AY: {en_iyi_ay.upper()} (Başarı Oranı: %{en_yuksek_oran:.1f})")
```

### 📞 "Bıkkınlık" Etkisi Analizi: Bir Müşteriyi En Fazla Kaç Kere Aramalıyız?

Pazarlamada çok ince ve tehlikeli bir çizgi vardır: Müşteriyi **ikna etmek** ile **bıktırmak** arasındaki sınır. 

Veri setimizdeki `campaign` sütunu, bu kampanya süresince bir müşterinin banka tarafından kaç defa arandığını/iletişime geçildiğini gösteriyor. Veri Bilimcisi (Data Scientist) şapkasıyla şu kritik soruyu soruyoruz: *"Bir insanı 5 kere, 10 kere hatta 20 kere aramak onun ikna olup vadeli hesap açtırma ihtimalini artırır mı? Yoksa belli bir aramadan sonra müşteri bıkıp teklifi tamamen reddeder mi?"*

Eğer başarı oranı belli bir aramadan sonra dramatik şekilde düşüyorsa (Azalan Verimler Kanunu), banka çağrı merkezine bir **üst limit (Threshold)** koymalıdır. Böylece hem çağrı merkezi personelinin değerli zamanı boşa harcanmaz hem de müşterinin markaya olan nefreti tetiklenmez.

Aşağıdaki analizde arama sayılarını mantıksal gruplara ('1 Kere', '2 Kere', ..., '10+ Kere') bölecek ve arama sayısı arttıkça başarı oranının nasıl eridiğini göreceğiz. 👇


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. ARAMA SAYILARINI KATEGORİLEŞTİRME FONKSİYONU
# Grafikte 40-50 kere arananlar yüzünden kalabalık olmaması için grupluyoruz
def arama_grupla(x):
    if x == 1: return '1 Kere'
    elif x == 2: return '2 Kere'
    elif x == 3: return '3 Kere'
    elif x <= 5: return '4-5 Kere'
    elif x <= 10: return '6-10 Kere'
    else: return '10+ Kere'

# Yeni sütunumuzu oluşturuyoruz
df['Arama_Grubu'] = df['campaign'].apply(arama_grupla)
arama_sirasi = ['1 Kere', '2 Kere', '3 Kere', '4-5 Kere', '6-10 Kere', '10+ Kere']

# 2. GÖRSELLEŞTİRME KISMI
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Arama_Grubu', hue='deposit', order=arama_sirasi, palette=['#ff9999','#66b3ff'])

plt.title("Kampanya Süresince Müşteriyi Kaç Kere Aradık?", fontsize=14, fontweight='bold')
plt.ylabel("Kişi Sayısı", fontsize=11)
plt.xlabel("Arama (İletişim) Sayısı", fontsize=11)
plt.legend(title='Hesap Açtırdı mı?')
plt.tight_layout()
plt.show()


# 3. İSTATİSTİKSEL KANIT: BIKKINLIK DÜŞÜŞÜ
arama_toplam = df.groupby('Arama_Grubu').size()
arama_evet = df[df['deposit'] == 'yes'].groupby('Arama_Grubu').size()

# Oran hesaplaması (Sıfıra bölme hatası almamak için fillna ekliyoruz)
arama_basari = (arama_evet / arama_toplam * 100).fillna(0).reindex(arama_sirasi)

print("############### BIKKINLIK (FATIGUE) VE LİMİT RAPORU ###############")
print("Banka aynı müşteriyi üst üste aradığında başarı (kabul) oranı nasıl değişiyor?\n")

for grup, oran in arama_basari.items():
    print(f"- {grup} aranan grupta teklifi KABUL ETME oranı: %{oran:.1f}")

print("\n" + "-"*65)
print("📌 ÇIKARIM: Arama sayısı arttıkça başarı oranı hızla düşmektedir (Bıkkınlık Etkisi).")
print("Çağrı merkezi için optimum arama limiti (Threshold) belirlenmeli ve")
print("bu limitin üzerindeki aramalar 'Zaman/Maliyet İsrafı' olarak değerlendirilmelidir.")
```
